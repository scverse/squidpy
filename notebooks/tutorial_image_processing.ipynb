# Image (pre-)processing: smoothing

`process_img` is a general image processing function that wraps `skimage` functions. Currently, only `skimage.filters.gaussian` is supported (but extension is easily possible)

Here, we show how to use this function to smooth the high-resolution tissue image.


```python
import importlib
import squidpy.im
importlib.reload(squidpy.im)
```

```python
# import modules
import matplotlib.pyplot as plt
import os
import squidpy as sq
```

## Load data

```python
# set path to dataset
BASE_PATH = "../../data"
dataset_folder = os.path.join(BASE_PATH, "20191205_10XVisium_MouseBrainCoronal_giovanni.palla")
# load data
img = sq.im.ImageContainer(os.path.join(dataset_folder, 'V1_Adult_Mouse_Brain_image.tif'))
```

```python
plt.imshow(img.crop_corner(4000,4000,250,250)['image'])
```

## Smooth image
Gaussian smoothing with sigma 2

```python
sq.im.process_img(img, img_id='image', processing='smooth', processing_kwargs={'sigma': 2})
```

the smoothed image is saved in ``image_smooth``

```python
plt.imshow(img.crop_corner(4000,4000,250,250)['image_smooth'])
```
